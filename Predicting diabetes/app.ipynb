```python
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv('diabetes.csv')

inputData = data.iloc[:, 0:8]  
outputData = data.iloc[:, 8]    
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Input

model = Sequential([
    Input(shape=(8,)),  
    Dense(16, activation='relu'),
    Dense(12, activation='relu'),
    Dense(8, activation='relu'),
    Dense(8, activation='relu'),
    Dense(4, activation='relu'),
    Dense(2, activation='relu'),
    Dense(1, activation='sigmoid')
])
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

model.fit(inputData, outputData, epochs=300, batch_size=10)
```

```python
accuracy = model.evaluate(inputData, outputData,verbose=0)

print(f'Accuracy: {accuracy[1]*100}')
```

```python
predictions = (model.predict(inputData) >0.5).astype("int32")
```
